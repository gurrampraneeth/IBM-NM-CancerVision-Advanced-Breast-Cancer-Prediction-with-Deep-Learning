BIVARIATE ANALYSIS

```
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
url = "/content/House Price India.csv"
tips = pd.read_csv(url)

# Bivariate analysis using scatter plot
plt.scatter(tips['Date'], tips['grade of the house'])
plt.xlabel('Date')
plt.ylabel('grade of the house')
plt.title('Date vs grade of the house')

# Adding regression line
z = np.polyfit(tips['Date'], tips['grade of the house'], 1)
p = np.poly1d(z)
plt.plot(tips['Date'],p(tips['Date']),"r--")

# Displaying the plot
plt.show()
```
